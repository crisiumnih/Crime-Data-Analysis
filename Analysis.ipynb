```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('crimedata.csv')
```

```python

data
```

```python
data.reset_index()
len(data.index)
```

```python
list(data.columns)
```

```python
data = data.drop(['Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4'],axis = 'columns')
sorted_data = data.sort_values(by=['Part 1-2','DR_NO'])
sorted_data.reset_index(drop=True, inplace=True)
sorted_data
```

# No. of Crimes to timeline

```python
date_split = sorted_data['DATE OCC'].str.split(' ',expand =True).get(0)
month_split = date_split.str.split('/')
months = [mo[0] for mo in month_split]
year = [yo[2] for yo in month_split]
months
data_filter = sorted_data[['DR_NO','AREA','AREA NAME','Rpt Dist No','Part 1-2','Crm Cd',
                  'Crm Cd Desc','Vict Age','Vict Sex','Vict Descent']]
```

```python
data_filter['Month']=months
data_filter['Year']=year
data_filter.head()
```

```python
data_filter
```

```python
timeline = data_filter.sort_values(by=['Year','Month'])
timeline.reset_index(drop=True, inplace=True)
timeline['Crimes per month'] = timeline.groupby(['Year', 'Month'])['Year'].transform('count')
timeline.head()
```

```python
grouped = timeline.groupby('Year')
for year, group in grouped:
    plt.plot(group['Month'], group['Crimes per month'], label=year)
plt.xlabel('Month')
plt.ylabel('Crimes per month')
plt.title('Crimes per month Over Time')
plt.legend(loc='upper right')

plt.show()
```

# Crimes in an area

```python
area_count = data_filter['AREA NAME'].nunique()
area_count
```

```python
area_data = data_filter
area_data['Crimes in the resp Area'] = area_data.groupby('AREA NAME')['AREA NAME'].transform('count')
area_data
```

```python
area_crime_count = area_data.groupby('AREA NAME')['Crimes in the resp Area'].first()
area_crime_count.plot(kind='bar')
plt.xlabel('Area')
plt.ylabel('No. of Crimes')
plt.title('No. of Crimes by Area')
```

# Crime type

```python
Code_count = data_filter['Crm Cd'].nunique()
Code_count
```

```python
Code_data = data_filter
crime_cou = Code_data.groupby('Crm Cd Desc').size().reset_index(name='count')
crime_cou
```

```python
top_crimes = crime_cou.sort_values('count',ascending = False).head(15)
plt.figure(figsize=(8, 8))
plt.title('Top 15 Crimes by Type')
plt.pie(top_crimes['count'], labels=top_crimes['Crm Cd Desc'], autopct='%1.1f%%')
plt.show()
```

```python
top_crimes = crime_cou.sort_values('count',ascending = False).head(12)

total_cou = crime_cou['count'].sum()
other = total_cou - top_crimes['count'].sum()
other_row = pd.DataFrame({'Crm Cd Desc': ['Other'], 'count': [other]})
top_crimes = pd.concat([top_crimes, other_row], ignore_index=True)
plt.figure(figsize=(10, 10))
plt.title('Top 12 Crimes by Type')
plt.pie(top_crimes['count'], labels=top_crimes['Crm Cd Desc'], autopct='%1.1f%%')
plt.show()
```

# Status of Crimes

```python
stat = sorted_data['Status Desc'].unique()
stat_count = sorted_data['Status Desc'].value_counts().reset_index()
stat_count.columns = ['Status', 'Count']
```

```python
resolved = stat_count.loc[stat_count['Status'].isin(['Adult Other', 'Adult Arrest','Juv Arrest','Juv Other']), 'Count'].sum()
unres = stat_count.loc[stat_count['Status'].isin(['Invest Cont', 'UNK']), 'Count'].sum()
statusun = pd.DataFrame({'Status': ['Resolved', 'Unresolved'], 'Count': [resolved, unres]})
plt.figure(figsize=(8, 6))
plt.bar(statusun['Status'], statusun['Count'], color=['blue', 'orange'])
plt.title('Status of Crimes')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```
